```python
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
from datetime import date, datetime
from kaggle.api.kaggle_api_extended import KaggleApi
import zipfile
```

```python
api = KaggleApi()
api.authenticate()
api.dataset_download_file('ealaxi/banksim1', file_name='bs140513_032310.csv')

with zipfile.ZipFile('bs140513_032310.csv.zip', 'r') as zipref:
    zipref.extractall()
```

```python
archive_csv = pd.read_csv("C:/Users/rafael.marcelino/Documents/GitHub/PagSeguro_RM_FT/bs140513_032310.csv", header=0, quotechar="'")
```

```python
archive_csv["data_aux"] = None
archive_csv["data_aux"] = np.where((archive_csv["step"] >= 0) & (archive_csv["step"] <= 30), date(2021,1,1), archive_csv["data_aux"])
archive_csv["data_aux"] = np.where((archive_csv["step"] >= 31) & (archive_csv["step"] <= 60), date(2021,2,1), archive_csv["data_aux"])
archive_csv["data_aux"] = np.where((archive_csv["step"] >= 61) & (archive_csv["step"] <= 90), date(2021,3,1), archive_csv["data_aux"])
archive_csv["data_aux"] = np.where((archive_csv["step"] >= 91) & (archive_csv["step"] <= 120), date(2021,4,1), archive_csv["data_aux"])
archive_csv["data_aux"] = np.where((archive_csv["step"] >= 121) & (archive_csv["step"] <= 150), date(2021,5,1), archive_csv["data_aux"])
archive_csv["data_aux"] = np.where((archive_csv["step"] >= 151) & (archive_csv["step"] <= 180), date(2021,6,1), archive_csv["data_aux"])

archive_csv
```

```python
_fact_merchant_kpi = archive_csv.groupby(["data_aux", "merchant"], as_index=False).agg(tpv=('amount','sum'), qtd_transacoes=('merchant','count'))
_fact_merchant_kpi
```

```python
archive_csv["gender"].where((archive_csv["gender"] == "M") | (archive_csv["gender"] == "M"), 'T', inplace=True)
archive_csv
```

```python
archive_csv["mes"] = 0
archive_csv["mes"] = np.where((archive_csv["step"] >= 0) & (archive_csv["step"] <= 30), 1, archive_csv["mes"])
archive_csv["mes"] = np.where((archive_csv["step"] >= 31) & (archive_csv["step"] <= 60), 2, archive_csv["mes"])
archive_csv["mes"] = np.where((archive_csv["step"] >= 61) & (archive_csv["step"] <= 90), 3, archive_csv["mes"])
archive_csv["mes"] = np.where((archive_csv["step"] >= 91) & (archive_csv["step"] <= 120), 4, archive_csv["mes"])
archive_csv["mes"] = np.where((archive_csv["step"] >= 121) & (archive_csv["step"] <= 150), 5, archive_csv["mes"])
archive_csv["mes"] = np.where((archive_csv["step"] >= 151) & (archive_csv["step"] <= 180), 6, archive_csv["mes"])

archive_csv
```

```python
fact_merchant_kpi = archive_csv.groupby(by=["mes", "merchant"], as_index=False).agg(tpv=('amount','sum'), qtd_transacoes=('merchant','count'))
fact_merchant_kpi
```

```python
dim_merchant_category = archive_csv.groupby(["merchant", "category"], as_index=False).any()[["merchant", "category"]]
dim_merchant_category.rename(columns={'merchant':'id_merchant'}, inplace=True)
dim_merchant_category
```
